```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../processed/data.csv', delimiter=';')
df_synopsis = pd.read_csv('../src/data.csv', delimiter=';')
```

```python
final_df = df.merge(
  df_synopsis, on='tconst', how='left')
```

```python
final_df = final_df[final_df['synopsis'].notnull()]
```

```python
final_df['synopsis'] = final_df['synopsis'].apply(lambda synopsis: synopsis.strip().replace('"', ""))
```

```python
final_df.to_csv('data.csv', index=False, sep=';')
```
